## Suggested Citation

If you use or adapt material from this book, please cite it as an online book.  
A recommended citation format is:

> Quarles, B. (Year). *Introductory Physics*. East Texas A&M University.  
> Jupyter Book. URL

For example (APA-style):

> Quarles, B. (2025). *Introductory Physics*. East Texas A&M University.  
> Jupyter Book. https://github.com/saturnaxis/IntroPhysics

If you are citing a specific chapter, section, figure, or exercise, include the section title and provide the direct URL when possible.

Because this book is a living document, you may optionally include the **access date** in your citation.

## BibTeX Entry

```bibtex
@book{QuarlesIntroPhysics,
  author    = {Quarles, Billy},
  title     = {Introductory Physics},
  year      = {2025},
  publisher = {East Texas A\&M University},
  note      = {Jupyter Book},
  url       = {https://github.com/saturnaxis/IntroPhysics}
}
```
