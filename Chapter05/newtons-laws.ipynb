# Newton's Laws of Motion

## Forces

The study of motion is called [*kinematics*](https://en.wikipedia.org/wiki/Kinematics), which describes the way objects move through their velocity and acceleration.  

The study of what *causes objects to move* is called [**dynamics**](https://en.wikipedia.org/wiki/Dynamics_(mechanics)), which describes the motion of objects and systems through forces.  The foundation of dynamics are the laws of motion developed by Isaac Newton, although some of the fundamental principles were developed earlier by Galileo Galilei.

The development of Newton's laws marked a transition from Renaissance thinking (i.e., reliance on ancient texts) to Scientific Revolution thinking which developed many of the fundamental pillars of the modern era.  Newton's description of mechanics is now described as *classical mechanics* where it applies to most situations until the 20th century. The development of modern physics was prompted by discoveries of subatomic particles which move at speeds comparable to the speed of light.  To describe the motions of these particles, we need *quantum mechanics*.

### Working Definition of Force
We need a "working" definition of a force to help build some intuition until we get to the more formal definition given by Newton's 2nd Law.  An intuitive definition of a force is

- a push or a pull.

A push or a pull has both magnitude (i.e., how hard) and a direction, so it is a vector quantity.  Force can be represented by vectors or expressed as a multiple of a standard force.

Our everyday experience gives a good idea of how multiple forces add.  If 2 people push in different directions on a third person (see Fig. {numref}`{number}<force-vectors-fig>`), we might expect the total force to be in the direction of the resultant vector (or force).  Forces, like vectors, are represented by arrows.

### Vector Notation for Force

The SI unit of force is called the **Newton** (abbreviated $\rm N$), which is the force needed to accelerate an object with a mass of $1\ {\rm kg}$ at a rate of $1\ {\rm m/s^2}$ (i.e., $1\ {\rm N} = 1\ {\rm kg \cdot m/s^2}$).  *Note that a small apple has a weight of about $1\ \rm N$.*

We can describe a force in the form

\begin{align*} 
\vec{F} &= a\hat{i} + b\hat{j}, &\quad \text{(in 2D)} \\
\vec{F} &= a\hat{i} + b\hat{j} + c\hat{k}. &\quad \text{(in 3D)}
\end{align*}

In Fig. {numref}`{number}<force-vectors-fig>`, let's suppose that ice skater 1 (on the left) pushes horizontally with a force of $30\ {\rm N}$ to the right (or $\vec{F}_1 = 30 \hat{i}\ {\rm N}$).  Similarly, if ice skater 2 (on the bottom) pushes with a force of $40\ {\rm N}$ in the postive vertial direction, we would write $\vec{F}_2 = 40\hat{j}\ {\rm N}$.  

The resultant force causes the third ice skater to accelerate.  The **net external force** $\vec{F}_{\rm net}$ is found by taking the vector sum of **all external forces** acting on an object or system.  We can then represent the net extrnal force (in general) as:

\begin{align}
\vec{F}_{\rm net} &= \sum \vec{F}_{\rm ext}, \\
&= \vec{F}_1 + \vec{F}_2 + \cdots + \vec{F}_N, \\
& = F_{{\rm net},x}\hat{i} + F_{{\rm net},y}\hat{j} + F_{{\rm net},z}\hat{k}, \\
&= \sum_{i=1}^N F_{ix}\hat{i} + \sum_{i=1}^N F_{iy}\hat{j} + \sum_{i=1}^N F_{iz}\hat{k}.
\end{align}

In this example, we have $\vec{F}_{\rm net} = \vec{F}_1 + \vec{F}_2 = 30\hat{i} + 40\hat{j}\ {\rm N}$.  To find the magnitude of the net (resultant) force $F_{\rm net}$, we take sum the squares of the components and take the square root, or

$$ F_{\rm net} = \sqrt{(30\ {\rm N})^2 + (40\ {\rm N})^2} = 50\ {\rm N}, $$

where the direction is given by

$$ \theta = \tan^{-1}\! \left(\frac{F_{net,y}}{F_{net,x}}\right) = \tan^{-1}\! \left(\frac{40}{30}\right) = 53^\circ $$

measured from the positive $x$-axis, as shown in the free-body diagram in Fig. {numref}`{number}<force-vectors-fig>`b.

---
Let's suppose the ice skaters now push the third ice skater with $\vec{F}_1
 = 3\hat{i} + 8\hat{j}\ {\rm N}$ and $\vec{F_2} = 5\hat{i} + 4\hat{j}\ {\rm N}.$

 - **What is the resultant of these two forces?**

 We recognize that force is a vector, and therfore, we must add using the rules for vector additon:

 \begin{align*}
 \vec{F}_{\rm net} &=  \sum_{i=1}^2 F_{ix}\hat{i} + \sum_{i=1}^2 F_{iy}\hat{j},\\
 &= (3 + 5)\hat{i} + (8 + 4)\hat{j}, \\
  &= 8\hat{i} + 12\hat{j}\ {\rm N}.
 \end{align*}

## Newton's 1st Law

Newton's 1st law is a restatment of Galileo's principle of [**inertia**](https://en.wikipedia.org/wiki/Inertia), which states that an object in motion tends to slow down unless some effort is made to keep it moving.  **Newton's 1st law** give a deeper explanation of this observation.

```{admonition} Newton's 1st Law of Motion
:class: note

A body at rest remains at rest or, if in motion, remains in motion at constant velocity unless acted on a by a net external force.
```

Note the expression "constant velocity", which means that the object maintains a path along a straight line (since neither the magnitude nor direction changes).  Also the verb "remains" tells us that the keeps on moving (or not moving) as part of the status quo.

Newton's 1st law sayst that there must be a cause for any change in velocity (i.e., acceleration) to occur.  The cause is a net external force.  An object sliding across a table or floor slows down due to the net force of friction acting on the object.  **If friction disappears, will the object slow down?**

Consider an air hockey table (Fig. {numref}`{number}<hockey-puck-fig>`).  When the air is turned off, the puck slides only a short distance before friction slows it to a stop.  However, with the air turned on, it creates a *nearly* frictionless surface, and the puck glides a long distance without slowing down.  If we know enough about the friction, we can accurately predict how quickly the object slows down.

Newton's 1st law is general and can be applied to anything from an object sliding on table, a satellite in orbit, or even blood pumped from the heart.  Experiments have verified that nay change in velocity (speed or direction) must be caused by an external force.

Aristotle had claimed that objects had a natural state and needed little "angels" to keep the objects moving.  Tt was Galileo who questioned this idea, which led whim to ask the fundamental question: "What is the cause?"  Thinking in terms of *cause and effect* is fundamentally different from the typical ancient Greek approach. 

### Gravitation and Inertia
Regardless of the scale of an object, two properties remain valid: gravitation and inertia because they are both connected to mass.

- **Mass** is a measure of the amount of matter in something.  It has volume and takes up space.

- **Gravitation** is the attraction of one mass to another.  The magnitude of this attration is your weight, and it is a force.

Mass is related to **inertia**, which is the ability of an object to *resist changes in its motion*, or to resist acceleration.  Newton's 1st law is often called the **law of inertia**.

### Intertial Reference Frames
Newton's 1st law can also be state as *"Every body remains in its state of uniform motion in a straight line unless it is compelled to change that state by forces acting on it."*  

- To Newton, "uniform motion in a straight line" meant constant velocity, which includes the special case of zero velocity (or rest).

Therefore the 1st law says that the velocity of an object remains constant if the net force on it is zero, or

$$ \vec{F}_{\rm net} = \sum \vec{F}_{ext} = 0. $$

Newton's 1st law is also considered as a statement about reference frames.  It provides a method for identifying a special type of referece frame called the **inertial reference frame**.

In principle, we can make make the net force on a body zero, if its velocity relative to a given reference frame is constant and then that frame is defined as inertial (recall {numref}`relative_2D`).  An inertial reference frame is one where Newton's 1st law is valid.  

```{admonition} Inertial Reference Frame
:class: note

A reference frame moving at constant velocity relative to an inertial reference frame is also inertial.  A reference frame acclerating relative to an inertial frame is not inertial.
```

- *Are inertial frames common in nature?*

A reference frame at rest to the most distant (or "fixed") stars is inertial.  All frames moving uniformly with respect to this fixed-star frame are also inertial.  For example, a nonrotating reference frame attached to the Sun is (for all practical purposes) inertial because its velocity relative to the fixed stars does not vary by more than one part in $10^{10}$.

Earth accelerates relative to the fixed stars because it rotates on its axis and *revolves* around the Sun.  Hence a reference frame attached to its surface is not inertial.  For most problems, such a frame serves as a sufficently accurate *approximation* to an inertial frame because the acceleration relative to the fixed stars is small (${\sim}10^{-2}\ {\rm m/s^2}$).  Thus we consider reference frames fixed to the Earth as inertial.

No particular inertial frame is more special than any other.  All inertial frames are equivalent.  In analyzing a problem, we choose one particular inertial frame simply based on convenience.
 
### Newton's 1st Law and Equilibrium

Newton's first law tells us about the *equilibrium* of a system (i.e., $\vec{F}_{\rm net} = 0$).  From Fig. {numref}`{number}<force-vectors-fig>`, the forces combine to form a net external force: $\vec{F}_{\rm net} = \vec{F}_1 + \vec{F}_2$.  To create equilibrium, we require a *balancing force* that will produce a net force of zero.

- This force must be equal in magnitude, but opposite in direction to $\vec{F}_{\rm net}$ (i.e., $\vec{F}_{\rm bal} = -\vec{F}_{\rm net}$).

For the ice skaters, we found $\vec{F}_{\rm net} = 30\hat{i} + 40\hat{j}\, {\rm N}$, where the balancing force is 

\begin{align*}
\vec{F}_{\rm bal} &= -\vec{F}_{\rm net}, \\
&= -30\hat{i} - 40\hat{j}\, {\rm N}.
\end{align*}

Newton's 1st law in vector form is given as

\begin{align}
\vec{\rm v} = \text{constant, when } \vec{F}_{\rm net} = \vec{0}\ {\rm N}.
\end{align}

For the net force to equal zero, the velocity vector $\vec{\rm v}$ of the object is a constant.

```{admonition} Derivative of a Constant (Calculus-framing)
:class: note, dropdown

From calculus, we have the [constant rule](https://openstax.org/books/calculus-volume-1/pages/3-3-differentiation-rules) which means that for any function $f(x) = c$ (where $c$ is a constant), the derivative $f^\prime(x) = 0$.

*What does this mean for an object moving with constant velocity?*  We have loosely defined a force as a push or pull (or causing an object to accelerate).  Recall the definition of acceleration as

$$ \vec{a} = \frac{d\vec{v}(\rm t)}{dt}, $$

which means the acceleration is the slope of the velocity curve.  If $\vec{\rm v}(t) = c$, then the slope is equal to zero.

If there is no accleration, then there is no force (assuming mass remains constant).
```

A nonzero net force is required to change the state of motion of the car.  As a car moves with constant velocity, the friction force propels the car forward and opposes the drag force against it.

#### **Example Problem**: Applying Newton's 1st Law to your car
```{exercise}
:class: etamu-exercise

**The Problem**

> Newton's laws can be applied to all physical processes involving force and motion, including something as mundane as driving a car.
>
> (a) Your car is parked outside your house. Does Newton's first law apply in this situation? Why or why not?
>
> (b) Your car moves at constant velocity down the street. Does Newton's first law apply in this situation? Why or why not?

---

**The Model**

In both situations, the car is modeled as a single object moving in an inertial reference frame fixed to the ground. Newton's first law applies when the vector sum of the external forces acting on the object is zero, which implies zero acceleration.

In part (a), the car is at rest on level ground. Forces act on the car, but the car does not accelerate.

In part (b), the car moves at constant velocity on level ground. Forces act on the car, but its velocity does not change in either magnitude or direction.

In both cases, the question is whether the forces acting on the car balance so that the net force is zero.

---

**The Math**

Newton's first law states that if the net external force on an object is zero, then the object does not accelerate. This can be written as

$$\sum \vec{F} = \vec{0} \quad \Rightarrow \quad \vec{a} = \vec{0}.$$

An object with zero acceleration is either at rest or moving with constant velocity.

**(a)** Parked car

Choose a coordinate system with $+\hat{j}$ pointing upward. Two forces act on the car:

* The weight of the car, which acts downward and is written as $\overrightarrow{W} = -w\, \hat{j}$
* The force exerted by the ground on the car, which acts upward and is written as $\vec{N} = n\, \hat{j}$

The net force on the car is the vector sum of these forces,

$$\sum \vec{F} = \vec{N} + \overrightarrow{W}  = (n - w)\, \hat{j}.$$

Since the car is at rest and does not accelerate, Newton's first law requires that the net force be zero. This implies

$$n - w = 0,$$

so the upward force exerted by the ground balances the downward weight of the car.

---

**(b)** Car moving at constant velocity

Choose a coordinate system with $+\hat{i}$ in the direction of motion and $+\hat{j}$ upward. Forces act on the car both vertically and horizontally.

Vertically, the downward weight $\overrightarrow{W}  = -w\, \hat{j}$ and the upward force from the ground $\vec{N} = n\, \hat{j}$ must balance because the car **does not** accelerate vertically:

$$ n - w = 0.$$

Horizontally, a forward force from the road on the tires acts in the $+\hat{i}$ direction and is written as

$$\vec{F}_{\text{road}} = f\, \hat{i}.$$

A resistive force due to air drag and rolling resistance acts opposite the motion and is written as

$$\vec{F}_{\text{res}} = -f\, \hat{i}.$$

Constant velocity implies zero acceleration, so the net horizontal force must be zero:

$$\vec{F}_{\text{road}} + \vec{F}_{\text{res}} = \vec{0}.$$

With both the vertical and horizontal forces balanced, the total net force on the car is zero.

---

**The Conclusion**

(a) Yes. Newton's first law applies to a parked car because the forces acting on it balance so that the net force is zero. With zero net force, the acceleration is zero and the car remains at rest.

(b) Yes. Newton's first law also applies to a car moving at constant velocity. Although forces act on the car, they balance in both the vertical and horizontal directions, producing zero net force. With zero net force, the car does not accelerate and continues moving with constant velocity.

---

**The Verification**

We verify the qualitative force balance by assigning representative vector magnitudes and directions, summing the force vectors, and showing that the net force is zero in both scenarios. The resulting vector diagrams serve as free-body diagrams for each case.

```

`````python
import numpy as np
import matplotlib.pyplot as plt

w = 1.0
n = 1.0
f = 0.6

W_a = np.array([0.0, -w])
N_a = np.array([0.0, +n])
Fnet_a = W_a + N_a

W_b = np.array([0.0, -w])
N_b = np.array([0.0, +n])
F_road = np.array([+f, 0.0])
F_res = np.array([-f, 0.0])
Fnet_b = W_b + N_b + F_road + F_res

print("Scenario (a) net force:", Fnet_a)
print("Scenario (b) net force:", Fnet_b)
def draw_fbd(ax, forces, labels, note):
    ax.plot(-0.01, 0.0, "bo", markersize=7)
    for F, lab in zip(forces, labels):
        Fx, Fy = F[0], F[1]
        ax.annotate("", xy=(Fx, Fy), xytext=(0.0, 0.0),
                    arrowprops=dict(arrowstyle="->", lw=2))
        # Label placement logic
        if abs(Fx) > abs(Fy):  # horizontal force
            xlab = Fx + 0.2*np.sign(Fx)
            ylab = Fy
            ha, va = "center", "center"
        else:                  # vertical force
            xlab = Fx
            ylab = Fy + 0.15*np.sign(Fy)
            ha, va = "center", "center"
        ax.text(xlab, ylab, lab, fontsize=11, horizontalalignment=ha,verticalalignment=va)

    ax.set_aspect("equal", adjustable="box")
    ax.set_xticks([])
    ax.set_yticks([])
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.text(0.02, 0.95, note, transform=ax.transAxes, fontsize=12, va="top")
    ax.set_xlim(-2, 2)
    ax.set_ylim(-2, 2)

fig = plt.figure(figsize=(7, 3.8), dpi=120)

ax1 = fig.add_subplot(121)
draw_fbd(ax1, [N_a, W_a],[r"$\vec{N}$", r"$\overrightarrow{W}$"],r"Parked car: $\sum \vec{F}=\vec{0}$")
ax1.text(0.,-1.5,r"$v = 0\ {\rm km/h}$ ",fontsize=11, horizontalalignment='center',verticalalignment='center')

ax2 = fig.add_subplot(122)
draw_fbd(ax2, [N_b, W_b, F_road, F_res],[r"$\vec{N}$", r"$\overrightarrow{W}$",
          r"$\vec{F}_{\rm{road}}$", r"$\vec{F}_{\rm{res}}$"],r"Constant velocity: $\sum \vec{F}=\vec{0}$")
ax2.text(-0.5,-1.5,r"$v = 50\ {\rm km/h}$ ",fontsize=11, horizontalalignment='center',verticalalignment='center')
ax2.annotate("", xy=(1, -1.5), xytext=(0.25, -1.5),
                    arrowprops=dict(arrowstyle="->", lw=2))
plt.tight_layout()
plt.show()
`````

## Newton's 2nd Law
Newton's 2nd law gives the cause-and-effect relationship between force and changes in motion.  It is quantitative and used extensively to calculate what happens in situations involving a force.

### Force and Acceleration

*What do we mean by a change in motion?*

A change in motion is equivalent to a change in velocity (i.e., an acceleration).  Newton's 1st law says that a net external force *causes* a change in motoin, and thus, we see that a *net external force causes nonzero acceleration*.

- An external force is one that is outside the system of interest.  In Fig. {numref}`{number}<car-ext-force-fig>`a, the system of interest is teh car plus the person within it.  The two forces exerted by the students outside are external forces.

- An internal force cats between elements of the system.  Thus the force the person exerts holding onto the steering wheel is an internal force. 

```{figure-md} car-ext-force-fig
<img src="https://openstax.org/apps/image-cdn/v1/f=webp/apps/archive/20251118.192121/resources/f46a0117dbe1644ffa4bb198eed333839ddacc67" alt="external force on car"  style="max-width:70%; height:auto;">

Image Credit: Openstax
```

Only external forces affect the motion of a system, according to Newton's 1st law.  Therefore, we must define the boundaries of the system to separate the internal from external forces.

In Fig. {numref}`{number}<car-ext-force-fig>`a, there are arrows denoting the external forces present:

- The weight $\vec{w}$ includes the car + person inside.
- The car is supported by the ground $\vec{N}$, which will cancel the weight, since the car is not floating nor sinking.
- The friction $f$ resists the motion and must be oriented opposite (to the left).
- The forces from each student pushing ($\vec{F}_1$ and $\vec{F}_2$) are applied to accelerate the car to the right.

All the forces add togehter to produce the net force $\vec{F}_{\rm net}$ (see Fig. {numref}`{number}<car-ext-force-fig>`b).  The free-body diagram (Fig. {numref}`{number}<car-ext-force-fig>`a) shows all of the forces acting on the system, where the dot at the center represents the center-of-mass.  Forces in the same direction are shown collinearly.

In Fig. {numref}`{number}<car-ext-force-fig>`c, a larger net external force produces a larger accleration $\left(\vec{a}^\prime > \vec{a}\right)$ when the tow truck pulls the car.  The acceleration is proportional to and in the same direction as the **net external force** action on a system.  Mathematically this is written as

$$ \vec{a} \propto \vec{F}_{\rm net}, $$

where the symbol $\propto$ (or `\propto`) means "proportional to."  

Once the system of interest is chosen, indentify the external force and ignore the interal ones.  It is a huge simplification to disregard the numerous internal forces acting between objects within the system.  This simplificaiton helps us solve some complex problems.

Recall that inertia is the "resistance to motion" and is related to an objects mass.  Therefore it seems reasonable that a more massive object requries more acceleration for it to move.  

In Fig. {numref}`{number}<fbd-force-fig>`, the same net external force applied to a basketball produces a much smaller acceleration when it is applied to an SUV.

```{figure-md} fbd-force-fig
<img src="https://openstax.org/apps/image-cdn/v1/f=webp/apps/archive/20251118.192121/resources/44230a76249c989838ed3776b13f50f15a74690c" alt="free-body diagrams for force"  style="max-width:70%; height:auto;">

Image Credit: Openstax
```

Therefore, we can write the proportional relationship between acceleration and mass as

$$ a \propto \frac{1}{m}, $$

where $m$ is the mass of the system and $a$ is the *magnitude* of the acceleration.  Combining the two proportionalities yields **Newton's 2nd law.**

```{admonition} Newton's 2nd Law of Motion
:class: note

The acceleration of a system is 

- directly proportional to the external force acting on the system, 
- in the same direction as the external force action on the system, and 
- inversely proportional to its mass.

In equation form, Newton's 2nd law is

\begin{align}
\vec{a} &= \frac{\vec{F}_{\rm net}}{m}, \\ \\
\vec{F}_{\rm net} &= \sum \vec{F} = m\vec{a}.
\end{align}

Since the mass $m$ is a scalar, we can see that it scales the vector $\vec{a}$ to give the net external force $\vec{F}_{\rm net}$.  Also we can use the properties of vectors to write the scalar equivalent as

\begin{align}
F_{\rm net} = ma.
\end{align}
```

**The Model**

Model the lawn mower as a particle moving on level ground. The problem states the net external force is already known and is horizontal, so we do not need to model the individual forces from the push and friction separately. ---

**The Math**

Newton's second law relates the net force on an object to its acceleration. In vector form,

$$
\sum \vec{F} = m\,\vec{a}.
$$

Choose $+\hat{i}$ to the right, parallel to the ground. The net force is given as a single horizontal vector,

$$
\vec{F}_{\text{net}} = (51\ \text{N})\,\hat{i}.
$$

Since the motion is along $\hat{i}$, write the acceleration as $\vec{a} = a\,\hat{i}$. Substitute into Newton's second law:

$$
\vec{F}_{\text{net}} = m\,\vec{a}
\quad \Rightarrow \quad
(51\ \text{N})\,\hat{i} = (24\ \text{kg})\,(a)\,\hat{i}.
$$

The unit vectors are the same on both sides, so the magnitudes must satisfy

$$
51\ \text{N} = (24\ \text{kg})\,a.

Solve for $a$:

$$
a = \frac{51\ \text{N}}{24\ \text{kg}}.

Using $1\ \text{N} = 1\ \text{kg}\cdot \text{m/s}^2$, the units reduce to $\text{m/s}^2$:

**The Conclusion**

The lawn mower's acceleration is $2.1\ \text{m/s}^2$ in the same direction as the net force, which is parallel to the ground.

Newton's first law states that if the net external force on an object is zero, then the object moves with constant velocity. Newton's second law states that a nonzero net external force produces acceleration. In vector form,

$$
\sum \vec{F} = \vec{0} \quad \text{(no acceleration)}

For the net force to point in the $+\hat{i}$ direction, the forward friction force must have a greater magnitude than the backward drag force.

**The Conclusion**

(a) When the car moves at constant speed, the net horizontal force is zero. Therefore, $\vec{F}_{\text{friction}}$ and $\vec{F}_{\text{drag}}$ have equal magnitudes and opposite directions.

(b) When the car accelerates to the right, the net horizontal force points to the right. Therefore, the magnitude of $\vec{F}_{\text{friction}}$ is greater than the magnitude of $\vec{F}_{\text{drag}}$.

###  Component Form of Newton's 2nd Law
Newtons 2nd law can be written in vector form as

$$ \sum \vec{F}_{\rm ext} = m\vec{a}. $$

Writing this equation in component form gives

\begin{align}
\sum F_x &= m a_x, \\
\sum F_y &= m a_y, \\
\sum F_z &= m a_z .
\end{align}

Once we write Newton's 2nd law in component form, the quantities $F_x$, $F_y$, $F_z$, $a_x$, $a_y$, and $a_z$ are scalars, so vector notation is no longer used.  Each component equation relates forces and acceleration along a single axis and can be analyzed independently.

The second law is a description of how a body responds *mechanically* to its environment.  The influence of the environment is the net force $\vec{F}_{\rm net}$, the body's response is the acceleration $\vec{a}$, and the strength of the response is inversely proportional to the mass $m$.  The larger the mass of an object, the smaller its acceleration to a given net force.

Model the ball as a particle. The given acceleration is already written in $\hat{i}$ and $\hat{j}$ form, so we apply Newtons second law directly in vector form. The net force on the ball is related to the acceleration by

$$

**The Math**

Because the acceleration vector is already expressed in terms of its $x$- and $y$-components, we can multiply each component by the mass to obtain the corresponding components of the net force.

$$ \vec{F}_{\rm net} = (0.400\ \text{kg})(3.00\hat{i} + 7.00\hat{j}\ \text{m/s}^2). $$

$$\vec{F}_{\rm net} = 1.20\hat{i} + 2.80\hat{j}\ \text{N}.$$

This expression answers part (a): the net force acting on the ball is a vector with both horizontal and vertical components.

To answer part (b), we determine the magnitude and direction of this force. The magnitude of a vector written in component form is found using the Pythagorean theorem:

$$ F_{\rm net} = \sqrt{F_x^2 + F_y^2}. $$

Substituting the force components,

$$ F_{\rm net} = \sqrt{(1.20\ \text{N})^2 + (2.80\ \text{N})^2} = 3.05\ \text{N}. $$

\end{align*}

Both descriptions represent the same physical net force.

\end{align*}

In the $x$-direction, the force $F_1$ contributes a component $F_1 \cos 30^\circ$ in the positive $x$-direction, while $F_3$ acts entirely in the negative $x$-direction. Thus,

\begin{align*}
F_1 \cos 30^\circ - F_3 &= m a_x .
\end{align*}

Substituting numerical values gives

\begin{align*}
(10.0\ \text{N})(\cos 30^\circ) - 5.0\ \text{N} &= (4.0\ \text{kg}), a_x, \\
a_x &= 0.92\ \text{m/s}^2 .
\end{align*}

In the $y$-direction, the upward forces are the vertical component of $F_1$ and the force $F_4$, while $F_2$ acts downward.
The force balance becomes

\begin{align*}
F_1 \sin 30^\circ + F_4 - F_2 &= m a_y .
\end{align*}

Substituting values,

\begin{align*}
(10.0\ \text{N})(\sin 30^\circ) + 2.0\ \text{N} - 40.0\ \text{N}
&= (4.0\ \text{kg}), a_y,\\
a_y &= -8.3\ \text{m/s}^2 .
\end{align*}

Combining the components, the acceleration vector is

\begin{align*}
\vec{a} = \left(0.92\,\hat{i} - 8.3\,\hat{j}\right)\ \text{m/s}^2 .
\end{align*}

---

**The Conclusion**

The particle accelerates with components $a_x = 0.92\ \text{m/s}^2$ and $a_y = -8.3\ \text{m/s}^2$.  This means the motion is predominantly downward, with a much smaller acceleration to the right.

The result reflects the fact that the large downward force $F_2$ dominates the vertical force balance, while the horizontal forces partially cancel. The direction and magnitude of the acceleration are entirely determined by the vector sum of the applied forces and the particles mass.

---

**The Verification**

The force components are recomputed numerically and divided by the mass to confirm the acceleration components obtained analytically.
````

`````python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Given values
# -----------------------------
m = 4.0
F1 = 10.0
F2 = 40.0
F3 = 5.0
F4 = 2.0
theta = np.deg2rad(30.0)

# -----------------------------
# Force components
# -----------------------------
F1_vec = np.array([F1*np.cos(theta), F1*np.sin(theta)])
F2_vec = np.array([0.0, -F2])
F3_vec = np.array([-F3, 0.0])
F4_vec = np.array([0.0, F4])

# Net force and acceleration
F_net = F1_vec + F2_vec + F3_vec + F4_vec
a_vec = F_net / m

# Sig-figappropriate output (2 sig figs)
print(f"Net force vector (N): [{F_net[0]:.2g}, {F_net[1]:.2g}]")
print(f"Acceleration vector (m/s^2): [{a_vec[0]:.2g}, {a_vec[1]:.3g}]")

`````

### Newton's 2nd Law and Momentum

Newton actually stated his 2nd law in terms of momentum

```{epigraph}
"The instantaneous rate at which a body's momentum changes is equal to the net force action on the body."
```

This is written by the vector equation:

\begin{align}
\vec{F}_{\rm net} = \frac{d\vec{p}}{dt}.
\end{align}

Momentum was described by Newton as "quantity of motion," or a way of combining both the velocity and mass of an object.  For now, it's sufficient to define *momentum* $\vec{p}$ as the product of the mass of the object $m$ and its velocity $\vec{\rm v}$:

$$ \vec{p} = m\vec{\rm v}. $$

Since velocity is a vector, so is momentum.  If we substitute our definition of momentum into Newton's 2nd law, we find

\begin{align*}
\vec{F}_{\rm net} &= \frac{d\vec{p}}{dt} = \frac{d(m\vec{\rm v})}{dt}, \\
&= m\frac{d{\vec{\rm v}}}{dt} = m\vec{a}.
\end{align*}

Thus, we see that the momentum form of Newton's 2nd law reduces to the form given earlier.

### Weight and Gravitational Force

When an object is dropped on Earth, it accelerates toward Earth's center.  Newton's 2nd law says that a net force on an object is responsible for its accleration.  If air resistance is negligible, the net force on a falling object is the gravitational force, which is called its **weight $\vec{w}$**, or its force due to gravity acting on an object of mass $m$.

Weight is deonted as a vector because it has a direction (*down*) and hence, weight is a downward force.  The magnitude of weight is denoted by $w$.  Using Newton's 2nd law, we can derive an equation for weight.

Consider an object with mass $m$ falling toward Earth.  It experiences only the downward force of gravity, which is the weight $\vec{w}$.  Newton's 2nd law says that the magnitude fo the net external force on an object is $\vec{F}_{\rm net} = m\vec{a}$.  We know that the acceleration due to gravity is $\vec{g}$, or $\vec{a} = \vec{g}$.

```{admonition}  Weight
:class: note

The gravitational force on a mass is its weight.  We can write this in vector form,

\begin{align}
\vec{w}  = m\vec{g},
\end{align}

where $m$ is the mass and $\vec{w}$ is the weight.  In scalar form, we can write

$$ w = mg. $$
```

The weight of a $1$-kg object on Earth is $9.81\ {\rm N}$:

$$ w = mg = (1.0\ {\rm kg})(9.81\ {\rm m/s^2}) = 9.81\ {\rm N}. $$

When the net external force on an object is its weight, we say that it is in **free fall**, which means that the only force acting on the object is gravity.

Accleration due to gravity $g$ varies slightly over Earth's surface, so the weight of an object depends on its location and is not an intrinsic property of the object.  Weight varies dramtically if we leave Earth's surface, where on the Moon $g\approx 1.62\ {\rm m/s^2}$ (about 1/6 of the value on Earth's surface).  A $1$-kg weight has a weight of $9.8\ {\rm N}$ on Earth, while it is only $1.6\ {\rm N}$ on the Moon.

```{note}
Weight and mass are different physical quantities, although they are closely related.  

- Mass is an intrinsic property of an object: it is a quantity of matter.  In most situations, mass does not vary; therefore its response to an applied force does not vary.
- Weight is the gravitational force acting on an object, so it does vary depending on gravity.  A person at a lower elevation in New Orleans (i.e., closer to Earth's center) weighs more than someone at high elevation like on top of Mt. Everest (i.e., farther from Earth's center) even though they have the same mass.
```

- *Why does if feel comfortable to equate mass and weight?*

On Earth, we see that weight $\vec{w}$ is directly proportional to mass $m$, since $\vec{w} = m\vec{g}$.  When two quantities are proportional, we develop [**heuristics**](https://en.wikipedia.org/wiki/Heuristic) to communicate more effectively among people with a shared sense of the heuristic.  

Suppose you are asked how far is Dallas, TX from Commerce, TX.  You might reply that it is about $1$-hr, depending on traffic.  This makes sense because you and your interlocutor both share in the assumption that you will travel by car at about $60\ {\rm mph}$.  It is important to note that time is **not equal** to distance, where you are communicating $x = vt$, where $x$ is proportional to $t$ because of the implicit assumption that $v$ is a constant.

The Earth rotates once every $24$-hr, which means that is about $15^\circ$ per hour.  Astronomers developed a measure of distance on the sky called [*right ascension*](https://en.wikipedia.org/wiki/Right_ascension) based on the *seemingly* constant rotation speed.  As a result, it carries units in hours, minute, seconds.


#### **Example Problem**: Lifting a Stone
```{exercise}
:class: etamu-exercise

**The Problem**

>A farmer is lifting some moderately heavy rocks from a field to pant crops.  He lifts a stone that weighs $40.0\ \text{lb}$ (about $180\ \text{N}$). What force does he apply if the stone accelerates upward at a rate of $1.5\ \text{m/s}^2$?

---

**The Model**

The system consists of the stone, treated as a particle moving only in the vertical direction. Two external forces act on the stone:

* the weight of the stone acting downward, and
* the upward force applied by the farmer.

The acceleration is upward, so the net force on the stone must also be upward. Air resistance is neglected.

---

**The Math**

The weight of the stone is given, but Newtons second law requires the mass.  We therefore begin by using the definition of weight,

$$ w = mg. $$

Solving for the mass gives

$$ m = \frac{w}{g}. $$

Substituting the given values,

$$ m = \frac{180\ \text{N}}{9.81\ \text{m/s}^2} = 18\ \text{kg}.$$

Next, we apply Newtons second law in the vertical direction.  Taking upward as the positive direction, the net force is the applied force minus the weight,

$$ \sum F_y = F - w. $$

Newtons second law gives

\begin{align*}
F - w &= ma, \\
F - 180\ \text{N} &= (18\ \text{kg})(1.5\ \text{m/s}^2), \\
F - 180\ \text{N} &= 27\ \text{N},
\end{align*}

so the applied force is

$$ F = 207\ \text{N} \approx  210\ \text{N}. \quad \text{(rounding 2 sig figs)}
$$

---

**The Conclusion**

The farmer applies an upward force of approximately $F = 210\ \text{N}.$

This force must be greater than the weight of the stone because the stone is accelerating upward. If the applied force were equal to the weight, the net force would be zero and the stone would move at constant velocity instead of accelerating.

This example reinforces that Newtons second law depends on the **net force**, not on any single force acting on an object, and that additional equations such as the definition of weight may be required to complete a dynamics analysis.

---

**The Verification**
The calculation is verified numerically by computing the mass from the given weight and then applying Newtons second law to recover the required applied force. A free-body diagram is also generated to confirm that the net force is upward, consistent with the stated upward acceleration.
```

`````python
import numpy as np
import matplotlib.pyplot as plt

# Given values
w = 180.0      # N
g = 9.81        # m/s^2
a = 1.5        # m/s^2

# Mass
m = w/g
# Applied force
F = m*a + w

print(f"Mass of stone = {m:.2g} kg")
print(f"Applied force = {F:.2g} N")

# --- Free-body diagram ---
def draw_fbd(ax, forces, labels, note):
    ax.plot(0.0, 0.0, "b.", markersize=17)
    for Fvec, lab in zip(forces, labels):
        Fx, Fy = Fvec
        ax.annotate("", xy=(Fx, Fy), xytext=(0.0, 0.0),arrowprops=dict(arrowstyle="->", lw=2))
        if abs(Fy) > abs(Fx):
            ax.text(Fx, Fy + 0.15*np.sign(Fy), lab,fontsize=14, ha="center", va="center")
        else:
            ax.text(Fx + 0.15*np.sign(Fx), Fy, lab,fontsize=14, ha="center", va="center")
    ax.set_aspect("equal")
    ax.set_xticks([])
    ax.set_yticks([])
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.set_xlim(-1, 1)
    ax.set_ylim(-2.5, 2.5)
    ax.text(0.02, 0.95, note, transform=ax.transAxes,
            fontsize=12, va="top")

# Scaled forces for plotting
forces = [np.array([0.0,  1.2]),   # applied force
          np.array([0.0, -1.0])    # weight
]

labels = [r"$\vec{F}$", r"$\vec{w}$"]

fig = plt.figure(figsize=(4, 4), dpi=120)
ax = fig.add_subplot(111)

draw_fbd(ax, forces, labels, r"Stone accelerating upward: $\sum \vec{F} = m\vec{a}$")
ax.text(0.25,-0.5, f"$w = %2i\ N$" % w,fontsize=14)
ax.text(0.25,0.5, f"$F = %2i\ N$" % F,fontsize=14)
ax.axhline(0,0,1,linestyle='--', color='k')
plt.tight_layout()
plt.show()
`````

## Newton's 3rd Law

```{admonition} Newton's 3rd Law of Motion
:class: note

Whenever one body exerts a force on a second body, the first body experiences a force that is equal in magnitude and opposite in direction.  Mathematically, if a body $A$ exerts a force $\vec{F}$ on body $B$, then $B$ simultaneously exerts a force $-\vec{F}$ on $A$, or in vector equation form,

\begin{align}
\vec{F}_{AB} = - \vec{F}_BA.
\end{align}
```

Newton's 3rd law represents a certain *symmetry* in nature: Forces always occur in pairs, and one body cannot exert a force on another without experiencing a force itself.  We sometimes refer to this law loosely as "action-reaction".

Consider a swimmer pushing off the side of a pool (see Fig. {numref}`{number}<swim-force-fig>`). She pushes against the pool wall with her feet and accelerates in the opposite direction of her push.  The wall has exerted an equal and opposite force on teh swimmer.

```{figure-md} swim-force-fig
<img src="https://openstax.org/apps/image-cdn/v1/f=webp/apps/archive/20251118.192121/resources/4e818036eb8843a742617e19afaaf980f3e5b9a4" alt="swim reaction force"  style="max-width:70%; height:auto;">

Image Credit: Openstax
```

You might think that two equal and opposite forces would cancel, but they do not because *they acto on different systems.*  In this case, there are two systems that we coul investigate: (a) the swimmer and (b) the wall.

- (a) If we select the swimmer, then $\vec{F}_\text{wall on feet}$ is an external force on this system and affects the swimmer's motion.  The swimmer moves in the direction of this force.
- (b) The force $\vec{F}_\text{feet on wall}$ acts on the wall and thus, it does not directly affect the motion of the swimmer and does not cancel $\vec{F}_\text{wall on feet}$.

The swimmer pushes in the direction opposite that in which she wishes to move.  The reactio to her push is thus in the desired direction.  In a free-body diagram, we never include both force of an action-reaction pair, we only use $\vec{F}_\text{wall on feet}$, not $\vec{F}_\text{feet on wall}$.

Other examples of Newton's 3rd law are easy to find:

- A car accelerates forward because the ground pushes forward on the drive wheels, in reaction to the drive wheels pushing backward on the gorund.
- Rockets move forward by expelling gas backward at high velocity.  This reaction force is called **thrust**, which pushes a body forward in responds to a backward force.
- Helicopters create lift by pushing air down, thereby experiencing an upward reaction force.

There are two important features of Newton's 3rd law.

1. The forces exerted (action and reaction) are always equal in magnitude but opposite in direction.
2. These forces are acting on different bodies or systems: $A$'s force acts on $B$ and $B$'s force on A.  They don't act on the same body and thus, they do not cancel each other.

A person who is walking or running applies Newton's 3rd law instinctively.  For example, a runner pushes backward on the ground so that it pushes him forward (see Fig. {numref}`{number}<runner-reaction-fig>`).

$$\sum \vec{F} = \vec{0}.$$

$$ \vec{S} + \vec{W} = \vec{0}. $$

$$ s\hat{j} - w\hat{j} = \vec{0}.$$

$$ (s - w)\hat{j} = \vec{0}. $$

This requires the scalar magnitudes to satisfy

$$ s = w. $$

The scale reading equals the magnitude of the gravitational force acting on the package. ````{exercise}
:class: etamu-exercise

**The Problem**

> A physics professor pushes a cart of demonstration equipment to a lecture hall (Figure {numref}`{number}<cart-reaction-fig>`). Her mass is $65\ {\rm kg}$, the carts mass is $12\ {\rm kg}$, and the equipments mass is $7\ {\rm kg}$. Calculate the acceleration produced when the professor exerts a backward force of $150\ {\rm N}$ on the floor. All forces opposing the motion, such as friction on the carts wheels and air resistance, total $24.0\ {\rm N}$.

```{figure-md} cart-reaction-fig
<img src="https://openstax.org/apps/image-cdn/v1/f=webp/apps/archive/20251118.192121/resources/7954635a55ec857e43667d5a90ed1aa42226b970" alt="cart reaction force"  style="max-width:70%; height:auto;">

Image Credit: Openstax
```

---

**The Model**

The system is the professor, the cart, and the equipment moving together as one object (System 1). Motion is horizontal, so we model the acceleration as one-dimensional along $\hat{i}$.

The floor exerts a forward force on the system because the professor pushes backward on the floor. Forces opposing the motion (wheel friction and air resistance) act backward on the system with a given total magnitude.

Vertical forces (weight and normal forces) balance so there is no vertical acceleration. Forces between the professor and the cart are internal to the chosen system and do not affect the net external force on the system.

---

**The Math**

Newtons second law relates the net external force on a system to its acceleration,

$$ \sum \vec{F}_{\rm ext} = m\vec{a}. $$

Because the motion is along the horizontal direction, we write the horizontal force balance in terms of scalar components,

$$ \sum F_x = m a_x. $$

The forward external force on the system is the force from the floor. We represent it as a vector in the $+\hat{i}$ direction,

$$ \vec{F}_{\rm floor} = F_{\rm floor}\hat{i}.$$

The forces opposing the motion act in the $-\hat{i}$ direction, so we write

$$\vec{f} = -f\hat{i}.$$

The net external force is the sum of the external forces along $\hat{i}$,

\begin{align*}
\sum \vec{F}_{\rm ext} &= \vec{F}_{\rm floor} + \vec{f}, \\
&= F_{\rm floor}\hat{i} - f\hat{i}, \\
&= (F_{\rm floor}-f)\hat{i}.
\end{align*}

This means the horizontal component satisfies

$$ F_{\rm floor}-f = m a_x. $$

The total mass of the system is the sum of the professor, cart, and equipment masses,

$$ m = 65.0\ {\rm kg} + 12.0\ {\rm kg}+ 7.0\ {\rm kg} = 84.0\ {\rm kg}.
$$

The given forces are $F_{\rm floor}=150\ {\rm N}$ and $f=24.0\ {\rm N}$, so the net external force magnitude is

$$ F_{\rm floor}-f = 150\ {\rm N} - 24.0\ {\rm N} = 126\ {\rm N}.
$$

Solving for the acceleration gives

$$ a_x = \frac{F_{\rm floor}-f}{m} = \frac{126\ {\rm N}}{84.0\ {\rm kg}} = 1.5\ {\rm m/s^2}.
$$

---

**The Conclusion**

The system (professor + cart + equipment) accelerates forward at $1.5 {\rm m/s^2}$. This result follows because the only horizontal external forces on the chosen system are the forward force from the floor and the backward resistive forces. Internal forces between the professor and the cart cancel within the system and do not affect the net external force.

---

**The Verification**

The verification recomputes the net force and acceleration numerically using the given masses and forces. A free-body diagram for System 1 is also drawn using the same `draw_fbd` function used in the earlier problems.
````


`````python
import numpy as np
import matplotlib.pyplot as plt

# Given values
m_prof = 65.0   # kg
m_cart = 12.0   # kg
m_eq = 7.0      # kg
F_floor = 150.0 # N
f = 24.0        # N

m = m_prof + m_cart + m_eq
F_net = F_floor - f
a = F_net / m

print(f"Total mass m = {m:.2g} kg")
print(f"Net external force F_net = {F_net:.3g} N")
print(f"Acceleration a = {a:.2g} m/s^2")

def draw_fbd(ax, forces, labels, note, chain_label=r"$\vec{T}$"):
    ax.plot(-0.01, 0.0, "bo", markersize=7)
    x0, y0 = 0.0, 0.0
    x_chain, y_chain = 0.0, 0.0
    for F, lab in zip(forces, labels):
        Fx, Fy = float(F[0]), float(F[1])
        if lab == chain_label:
            ax.annotate("", xy=(x_chain + Fx, y_chain + Fy), xytext=(x_chain, y_chain),arrowprops=dict(arrowstyle="->", lw=2))
            ax.text(x_chain + 0.5*Fx, y_chain + 0.5*Fy + 0.18, lab, fontsize=11,horizontalalignment="center")
            x_chain += Fx
            y_chain += Fy
            continue
        ax.annotate("", xy=(x0 + Fx, y0 + Fy), xytext=(x0, y0),arrowprops=dict(arrowstyle="->", lw=2))
        if abs(Fx) > abs(Fy):
            xlab = x0 + Fx + 0.2*np.sign(Fx)
            ylab = y0 + Fy
        else:
            xlab = x0 + Fx
            ylab = y0 + Fy + 0.15*np.sign(Fy)
        ax.text(xlab, ylab, lab, fontsize=11, horizontalalignment="center",
                verticalalignment="center")
    ax.set_aspect("equal", adjustable="box")
    ax.set_xticks([])
    ax.set_yticks([])
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.text(0.02, 0.95, note, transform=ax.transAxes, fontsize=12, va="top")
    ax.set_xlim(-2.0, 2.8)
    ax.set_ylim(-2.0, 2.0)

# FBD (System 1): include all forces shown in the diagram
F_floor_plot = 1.4
f_plot = 0.8
N_plot = 1.2
w_plot = 1.2

forces = [
    np.array([+F_floor_plot, 0.0]),
    np.array([-f_plot, 0.0]),
    np.array([0.0, +N_plot]),
    np.array([0.0, -w_plot]),
]
labels = [
    r"$\vec{F}_{\rm floor}$",
    r"$\vec{f}$",
    r"$\vec{N}$",
    r"$\vec{w}$",
]

fig = plt.figure(figsize=(5,4), dpi=120)
ax = fig.add_subplot(111)
draw_fbd(ax, forces, labels, r"System 1: $\sum \vec{F}_{\rm ext}=m\vec{a}$")
plt.tight_layout()
plt.show()
`````

:class: etamu-exercise

**The Problem**

> Calculate the force the professor exerts on the cart in Figure {numref}`{number}<cart-reaction-fig>`, using data from the previous example if needed.

---

**The Model**

We define the system of interest to be the cart plus the equipment (System 2). The professors push on the cart is now an external force on this system, so it appears in Newtons second law for System 2.

Motion is horizontal, so we model the acceleration as one-dimensional along $\hat{i}$. The forces opposing the motion (wheel friction and air resistance) are treated as a single external resistive force of magnitude $24.0\ {\rm N}$ acting in the $-\hat{i}$ direction. Vertical forces balance, so $a_y=0$.

---

**The Math**

Newtons second law in vector form is

$$ \sum \vec{F}_{\rm ext} = m\vec{a}. $$

For System 2, the horizontal external forces are the professors push on the cart and the resistive force. We write these forces as vectors along $\hat{i}$:

\begin{align*}
\vec{F}_{\rm prof} &= F_{\rm prof}\hat{i}, \\
\vec{f} &= -f\hat{i}.
\end{align*}

The net external force on System 2 is then

\begin{align*}
\sum \vec{F}_{\rm ext} &= \vec{F}_{\rm prof} + \vec{f}, \\
&= (F_{\rm prof}-f)\hat{i}.
\end{align*}

Since the acceleration is along $\hat{i}$, the scalar component equation is

$$F_{\rm prof}-f = m a.$$

We want $F_{\rm prof}$, so we solve for it:

$$F_{\rm prof} = m a + f.$$

The mass of System 2 is the cart plus the equipment,

$$m = 12.0\ {\rm kg} + 7.0\ {\rm kg} = 19.0\ {\rm kg}.$$

From the previous example, the acceleration of the whole group is

$$a = 1.5\ {\rm m/s^2}.$$

The resistive force is

$$f = 24.0\ {\rm N}.$$

Substituting gives

\begin{align*}
F_{\rm prof} &= (19.0\ {\rm kg})(1.5\ {\rm m/s^2}) + 24.0\ {\rm N}, \\
&= 28.5\ {\rm N} + 24.0\ {\rm N}, \\
&= 52.5\ {\rm N}.
\end{align*}

With significant figures based on the given measurements, the result is

$$ F_{\rm prof} = 53\ {\rm N}. $$

---

**The Conclusion**

The professor exerts a forward force of $53\ {\rm N}$ on the cart (System 2). This is smaller than the $150\ {\rm N}$ force involved at the floor in the previous example because that larger interaction accelerates the professor as well. Choosing a smaller system changes which forces are external, and it changes what force you are solving for.

---

**The Verification**

The verification recomputes $F_{\rm prof}=ma+f$ using the given masses and the previously found acceleration, and it draws the free-body diagram for System 2 showing $\vec{F}_{\rm prof}$ to the right and $\vec{f}$ to the left (with vertical forces included for completeness).

```

`````python
import numpy as np
import matplotlib.pyplot as plt

# Given values (from the previous example, as needed)
m_cart = 12.0     # kg
m_eq = 7.0        # kg
a = 1.5           # m/s^2
f = 24.0          # N

m = m_cart + m_eq
F_net = m*a
F_prof = F_net + f

print(f"System 2 mass m = {m:.3g} kg")
print(f"Net force needed (ma) = {F_net:.2g} N")
print(f"Professor's force on cart F_prof = {F_prof:.2g} N")

def draw_fbd(ax, forces, labels, note, chain_label=r"$\vec{T}$"):
    ax.plot(-0.01, 0.0, "bo", markersize=7)
    x0, y0 = 0.0, 0.0
    x_chain, y_chain = 0.0, 0.0

    for F, lab in zip(forces, labels):
        Fx, Fy = float(F[0]), float(F[1])
        if lab == chain_label:
            ax.annotate("", xy=(x_chain + Fx, y_chain + Fy), xytext=(x_chain, y_chain), arrowprops=dict(arrowstyle="->", lw=2))
            ax.text(x_chain + 0.5*Fx, y_chain + 0.5*Fy + 0.18, lab, fontsize=11, horizontalalignment="center")
            x_chain += Fx
            y_chain += Fy
            continue
        ax.annotate("", xy=(x0 + Fx, y0 + Fy), xytext=(x0, y0), arrowprops=dict(arrowstyle="->", lw=2))
        if abs(Fx) > abs(Fy):
            xlab = x0 + Fx + 0.2*np.sign(Fx)
            ylab = y0 + Fy
        else:
            xlab = x0 + Fx
            ylab = y0 + Fy + 0.15*np.sign(Fy)
        ax.text(xlab, ylab, lab, fontsize=11, horizontalalignment="center",verticalalignment="center")
    ax.set_aspect("equal", adjustable="box")
    ax.set_xticks([])
    ax.set_yticks([])
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.text(0.02, 0.95, note, transform=ax.transAxes, fontsize=12, va="top")
    ax.set_xlim(-2.0, 2.8)
    ax.set_ylim(-2.0, 2.0)

# --- Free-body diagram (System 2): include all forces shown for the system ---
F_prof_plot = 1.4
f_plot = 0.8
N_plot = 1.2
w_plot = 1.2

forces = [
    np.array([+F_prof_plot, 0.0]),
    np.array([-f_plot, 0.0]),
    np.array([0.0, +N_plot]),
    np.array([0.0, -w_plot]),
]

labels = [
    r"$\vec{F}_{\rm prof}$",
    r"$\vec{f}$",
    r"$\vec{N}'$",
    r"$\vec{w}'$",
]

fig = plt.figure(figsize=(5,4), dpi=120)
ax = fig.add_subplot(111)
draw_fbd(ax, forces, labels, r"System 2: $\sum \vec{F}_{\rm ext}=m\vec{a}$")
plt.tight_layout()
plt.show()
`````

## Common Forces

### Normal Force

Weight is a pervasive force that acts at all times and must be counteracted to keep an object from falling (i.e., in static equilibrium).  You must support the weight of a heavy object by pushing up on it when you hold it stationary.

When a bag of dog food is placed on the table (see Fig. {numref}`{number}<normal-force-fig>`), the table sags slightly under the load.  This is more noticeable on a card table, compared to a sturdy oak table.  Unless an object is deformed beyond its limit, *it will exert a restoring force* like a spring.  The greater the deformation, the greater the restoring force.  The next external force on the load is zero, when the load is stationary on the table. 

```{figure-md} normal-force-fig
<img src="https://openstax.org/apps/image-cdn/v1/f=webp/apps/archive/20251118.192121/resources/f15839d690c6a72056876d484e4d7cf88bfbc674" alt="normal force example"  style="max-width:70%; height:auto;">

Image Credit: Openstax
```

Whatever supports a load must supply an upward force *equal* to the weight of the load.  If the force supporting the weight of an object is perpendicular to the surface of contact, this is defined as a **normal force** $\vec{N}$.  This means that the normal force experienced by an object resting on a horizontal forced can be expressed in vector form and scalar form as:

\begin{align}
\vec{N} &= m\vec{g}, \\
N &= mg.
\end{align}

```{note}
One must be careful to not confuse the scalar magnitude of the normal force $N$ with the unit of force, the newton $\rm N$.

The normal force can be **less than** the object's weight, if the object is on an incline.
```

When an object rests on an incline that makes an angle $\theta$ with the horizontal, the force of gravity acting on the object is divided into two components:

1. a force acting perpendicular to the plane, $w_y$, and
2. a force acting parallel to the plane, $w_x$.

The normal force $\vec{N}$ is typically equal in magnitude and opposite in direction to $w_y$, where the component parallel to the plane $w_x$ causes the object to accelerate down the incline (see Fig. {numref}`{number}<incline-force-fig>`).  The $w_x$ is the unbalanced force which allows the accleration to occur, where $w_y$ is balanced by the normal force $N$.

```{figure-md} incline-force-fig
<img src="Inclined_Plane.png" alt="force on an incline example"  style="max-width:70%; height:auto;">

The coordinate system is chosen so that the $x$-axis is parallel to the inclined surface. The angle $\beta (= 90^\circ - \theta)$ is complementary to $\theta. Modified from OpenStax Figure 5.23.
```

```{admonition} Angles on a Triangle
:class: tip, dropdown

*How do we know that the angle $\theta$ given in Fig. {numref}`{number}<incline-force-fig>` is the same in both places?*

Let's assume that they are different: 

- $\theta_1$ is the angle between between $\vec{w}$ and $\vec{w}_y$,
- $\theta_2$ is the angle between the horizontal surface and incline plane.

The angle between $\vec{w}$ and $\vec{w}_x$ is $\beta = 90^\circ - \theta_1$ because $\vec{w}_x$ and $\vec{w}_y$ are perpendicular.  The inclined surface also makes a triangle, with 3 angles that sum to $180^\circ$, or

$$ \theta_2 + 90^\circ + \beta = 180^\circ.$$

We can solve for $\beta$ in terms of $\theta_2$ and substitute into the equation for the equation with $\theta_1$,

\begin{align*}
\beta &= 90^\circ - \theta_2, \\
90^\circ - \theta_2 &= 90^\circ - \theta_1,\\
\theta_2 &= \theta_1.
\end{align*}

With a little geometry, we showed that $\theta_1 = \theta_2 = \theta$.
```

**Be careful when resolving the weight into components.**  Normally when the $x$-component of a force $\vec{F}$ is resolved, we find that $F_x = F\cos{\theta}$.  However, in this case we are using the complementary angle $\beta = 90^\circ - \theta$ to resolve the $w_x$.  Using the red vectors in Fig. {numref}`{number}<incline-force-fig>`, we have

\begin{align*}
w_x &= w\cos{\beta} = w\sin{\theta}, \\
w_y &= w\cos{\theta}.
\end{align*}

We can sum the forces along the $y$-axis to write the magnitude of the normal force experienced by an object resting on an inclined plane:

\begin{align*}
\sum F_y &= N - mg\cos{\theta} = 0, \\
N &= mg\cos{\theta}.
\end{align*}

Notice that the force components and normal force could be determined from a free-body diagram.  

```{glue:figure} fbd_weight_components
:width: 560px
:align: center

Free-body diagram showing $\vec{w}$ resolved into $\vec{w}_x$ and $\vec{w}_y$ in axes aligned with the incline.
```


`````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Arc
from myst_nb import glue
# -----------------------------
# Parameters
# -----------------------------
theta = np.deg2rad(30.0)   # incline angle
w = 1.6                    # weight magnitude (plot units)
N = 1.4*np.cos(theta)                    # normal force magnitude (plot units)
# Weight components in rotated frame
w_x = w * np.sin(theta)
w_y = w * np.cos(theta)

# -----------------------------
# Helper: angle arc marker
# -----------------------------
def draw_angle_arc(ax, v1, v2, radius=0.6, label=r"$\theta$", label_r=0.75, lw=1.5):
    """
    Draw the smaller CCW arc from v1 to v2 (both are 2D direction vectors).
    Places a label near the middle of the arc.
    """
    def ang_deg(v):
        return (np.degrees(np.arctan2(v[1], v[0])) + 360.0) % 360.0
    a1 = ang_deg(v1)
    a2 = ang_deg(v2)
    # CCW sweep from a1 to a2
    delta = (a2 - a1) % 360.0
    if delta > 180.0:
        # Use the smaller arc by swapping
        a1, a2 = a2, a1
        delta = (a2 - a1) % 360.0
    a2_plot = a1 + delta  # ensure CCW arc length = delta
    arc = Arc((0.0, 0.0), 2*radius, 2*radius, angle=0.0, theta1=a1, theta2=a2_plot, lw=lw)
    ax.add_patch(arc)
    amid = np.radians(a1 + 0.5*delta)
    ax.text(label_r*np.cos(amid), label_r*np.sin(amid), label, fontsize=11,
            ha="center", va="center")
# -----------------------------
# Free-body diagram function
# -----------------------------
def draw_fbd(ax, forces, labels, note, colors=None):
    ax.plot(0, 0, "k.", markersize=15)
    if colors is None:
        colors = ["black"]*len(forces)
    for F, lab, col in zip(forces, labels,colors):
        Fx, Fy = F
        ax.annotate("",xy=(Fx, Fy),xytext=(0, 0),
            arrowprops=dict(arrowstyle="->", lw=2,color=col, shrinkA=0, shrinkB=0))
        # Label placement
        if abs(Fx) > abs(Fy):
            ax.text(Fx * 1.05-0.2, Fy-0.2, lab, color=col, ha="left", va="center", fontsize=14)
        else:
            ax.text(Fx+0.1, Fy * 1.08-0.025, lab, color=col, ha="center", va="center", fontsize=14)
    ax.set_aspect("equal", adjustable="box")
    ax.set_xticks([])
    ax.set_yticks([])
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.text(0.3, 0.95, note, transform=ax.transAxes, fontsize=12, va="top")

# -----------------------------
# Coordinate axes (rotated)
# -----------------------------
x_hat = np.array([np.cos(theta), np.sin(theta)])
y_hat = np.array([-np.sin(theta), np.cos(theta)])
# Directions for components (down-slope and into the plane-normal direction)
wx_dir = -x_hat
wy_dir = -y_hat
w_dir  = np.array([0.0, -1.0])  # weight direction (down)
# -----------------------------
# Forces
# -----------------------------
forces = [
    np.array([0.0, -w]),                        # weight
    np.array([N * y_hat[0], N * y_hat[1]]),     # normal force (along +y_hat)
    np.array([w_x * wx_dir[0], w_x * wx_dir[1]]),  # w_x (along -x_hat)
    np.array([w_y * wy_dir[0], w_y * wy_dir[1]])   # w_y (along -y_hat)
]
labels = [r"$\vec{w}$",r"$\vec{N}$",r"$\vec{w}_x$",r"$\vec{w}_y$"]
colors = ["k","k","r","r"]

# -----------------------------
# Plot
# -----------------------------
fig = plt.figure(figsize=(6, 4), dpi=120)
ax = fig.add_subplot(111)

draw_fbd(ax, forces, labels, r"Free-body diagram",colors=colors)
# --- Parallelogram construction (dashed lines) ---
w_tip  = forces[0]
wx_tip = forces[2]
wy_tip = forces[3]
ax.plot([wx_tip[0], w_tip[0]],[wx_tip[1], w_tip[1]],
        linestyle="--", linewidth=1.2, color="red")
ax.plot([wy_tip[0], w_tip[0]],[wy_tip[1], w_tip[1]],
        linestyle="--", linewidth=1.2, color="red")

# Draw coordinate axes
x_offset, y_offset = 0.2,0.1
ax.annotate("", xy=(x_hat[0]+x_offset, x_hat[1]+0.2), xytext=(x_offset, y_offset),
            arrowprops=dict(arrowstyle="->", lw=1.5,alpha=0.5))
ax.text(1.1*x_hat[0]+0.1, 1.1*x_hat[1], r"$x$", fontsize=11,alpha=0.5)

ax.annotate("", xy=(y_hat[0]+x_offset, y_hat[1]+0.1), xytext=(x_offset, y_offset),
            arrowprops=dict(arrowstyle="->", lw=1.5,alpha=0.5))
ax.text(1.1*y_hat[0]+x_offset, 1.1*y_hat[1]+y_offset/2, r"$y$", fontsize=11,alpha=0.5)
ax.plot(x_offset,0.1,'.',color='gray',ms=10)

# -----------------------------
# Angle markers
#   theta: between w and w_y direction (-y_hat)
#   beta : between w and w_x direction (-x_hat)
# -----------------------------
draw_angle_arc(ax, w_dir, wy_dir, radius=0.75, label=r"$\theta$", label_r=0.95)
draw_angle_arc(ax, wx_dir, w_dir, radius=0.45, label=r"$\beta$", label_r=0.60)

ax.set_xlim(-2., 2.)
ax.set_ylim(-2., 2.)

plt.tight_layout()
glue("fbd_weight_components", fig, display=False)
fig.savefig("inclined_FBD.png",bbox_inches='tight',dpi=150)
plt.show()
`````

#### **Example Problem**: Weight on an incline
````{exercise}
:class: etamu-exercise

**The Problem**

> Consider the skier on the slope in Figure {numref}`{number}<skier-incline-fig>`. Her mass including equipment is $60.0\ \text{kg}$.
> (a) What is her acceleration if friction is negligible? 
> (b) What is her acceleration if friction is $45.0\ \text{N}$?  

```{figure-md} skier-incline-fig
<img src="https://openstax.org/apps/image-cdn/v1/f=webp/apps/archive/20251118.192121/resources/095c5f6fefc5827b20d2c19a057c48ddb4e0d793" alt="skier incline example"  style="max-width:70%; height:auto;">

Image Credit: Openstax
```

---

**The Model**

The system of interest is the skier and her equipment, treated as a single particle of mass $m = 60.0\ \text{kg}$.  The skier moves on a rigid incline of fixed angle $\theta = 25^\circ$. Air resistance is neglected.

A coordinate system is chosen such that the $x$-axis is parallel to the slope and the $y$-axis is perpendicular to the slope. The forces acting on the skier are the gravitational force $\vec{w}$, the normal force $\vec{N}$, and (in part b) the friction force $\vec{f}$. There is no acceleration perpendicular to the slope.

---

**The Math**

Newtons second law is applied along each coordinate direction. Since there is no acceleration perpendicular to the slope, only forces parallel to the slope determine the motion.

The weight $\vec{w}$ is resolved into components parallel and perpendicular to the slope. The magnitudes of these components are

\begin{align*}
w_x &= w \sin\theta = mg\sin\theta, \\
w_y &= w \cos\theta = mg\cos\theta.
\end{align*}

**(a)** Negligible friction

When friction is negligible, the only force parallel to the slope is the component of weight $w_x$. Newtons second law along the slope gives

\begin{align*}
\sum F_x &= m a_x, \\
w_x &= m a_x.
\end{align*}

Substituting $w_x = mg\sin\theta$ and solving for the acceleration,

\begin{align*}
a_x &= \frac{mg\sin\theta}{m} = g\sin\theta.
\end{align*}

With $g = 9.81\ \text{m/s}^2$ and $\theta = 25^\circ$,

$$a_x = (9.81\ \text{m/s}^2)\sin 25^\circ = 4.1\ \text{m/s}^2.$$

**(b) Including friction**

When friction is present, it acts parallel to the slope and opposes the motion. The net force parallel to the slope is

\begin{align*}
F_{\text{net},x} &= w_x - f.
\end{align*}

Applying Newtons second law,

\begin{align*}
m a_x &= w_x - f = mg\sin\theta - f.
\end{align*}

Solving for the acceleration,

\begin{align*}
a_x &= \frac{mg\sin\theta - f}{m}, \\
&= g\sin{\theta} -\frac{f}{m}.
\end{align*}

Substituting numerical values,

$$a_x = (9.81\ \text{m/s}^2)\sin 25^\circ - \frac{45\ \text{N}}{60\ \text{kg}}
= 3.4\ \text{m/s}^2.$$

---

**The Conclusion**

With negligible friction, the skier accelerates down the slope at $4.1\ \text{m/s}^2$, determined entirely by the component of gravity parallel to the incline. When a friction force of $45.0\ \text{N}$ is present, the acceleration is reduced to $3.4\ \text{m/s}^2$ because friction opposes the downslope component of the skiers weight.

---

**The Verification**

We verify the accelerations by computing the component of the gravitational force parallel to the slope and applying Newtons second law numerically. A free-body diagram is also drawn using the same force components.

````

`````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Arc

# Given values
m = 60.0            # kg
g = 9.81            # m/s^2
theta = np.deg2rad(25.0)
f = 45.0            # N

# Weight components
w = m * g
w_x = w * np.sin(theta)
w_y = w * np.cos(theta)
# Accelerations
a_no_fric = w_x / m
a_with_fric = (w_x - f) / m
print(f"Acceleration without friction: {a_no_fric:.1f} m/s^2")
print(f"Acceleration with friction:    {a_with_fric:.1f} m/s^2")

# Normal force (just balanced visually)
N = w_y
# frictional force (just to make better visually on FBD)
f *= 3
# -----------------------------
# Helper: angle arc marker
# -----------------------------
def draw_angle_arc(ax, v1, v2, radius=0.6, label=r"$\theta$", label_r=0.75, lw=1.5):
    def ang_deg(v):
        return (np.degrees(np.arctan2(v[1], v[0])) + 360.0) % 360.0
    a1 = ang_deg(v1)
    a2 = ang_deg(v2)
    delta = (a2 - a1) % 360.0
    if delta > 180.0:
        a1, a2 = a2, a1
        delta = (a2 - a1) % 360.0
    arc = Arc((0, 0), 2*radius, 2*radius,
              theta1=a1, theta2=a1+delta, lw=lw)
    ax.add_patch(arc)
    amid = np.radians(a1 + delta/2)
    ax.text(label_r*np.cos(amid), label_r*np.sin(amid),
            label, ha="center", va="center", fontsize=11)
# -----------------------------
# Free-body diagram function
# -----------------------------
def draw_fbd(ax, forces, labels, note, colors=None):
    ax.plot(0, 0, "k.", ms=15)
    if colors is None:
        colors = ["black"] * len(forces)
    for F, lab, col in zip(forces, labels, colors):
        ax.annotate("", xy=F, xytext=(0, 0),
            arrowprops=dict(arrowstyle="->", lw=2,color=col, shrinkA=0, shrinkB=0))
        Fx, Fy = F
        if abs(Fx) > abs(Fy):
            ax.text(1.08*Fx-0.05, Fy+0.1, lab, color=col,
                    ha="center", va="center", fontsize=13)
        else:
            ax.text(Fx, 1.08*Fy-0.025, lab, color=col,
                    ha="center", va="center", fontsize=13)
    ax.set_aspect("equal")
    ax.set_xticks([])
    ax.set_yticks([])
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.text(0.05, 0.95, note, transform=ax.transAxes,
            fontsize=12, va="top")
# -----------------------------
# Rotated coordinate axes
# -----------------------------
x_hat = np.array([np.cos(theta), np.sin(theta)])
y_hat = np.array([-np.sin(theta), np.cos(theta)])
wx_dir = -x_hat
wy_dir = -y_hat
w_dir  = np.array([0.0, -1.0])
# -----------------------------
# Forces
# -----------------------------
plot_scale = 1 / w   # normalize weight to length 1
forces = [
    np.array([0.0, -w]),                  # weight
    N * y_hat,                             # normal
    w_x * wx_dir,                          # w_x
    w_y * wy_dir,                          # w_y
    f * x_hat                              # friction (upslope)
]
forces = plot_scale*np.array(forces)

labels = [r"$\vec{w}$",r"$\vec{N}$",r"$\vec{w}_x$",r"$\vec{w}_y$",r"$\vec{f}$"]
colors = ["k", "k", "r", "r", "b"]

# -----------------------------
# Plot
# -----------------------------
fig = plt.figure(figsize=(6, 4), dpi=120)
ax = fig.add_subplot(111)

draw_fbd(ax, forces, labels, "Free-body diagram", colors)

# Parallelogram construction
w_tip  = forces[0]
wx_tip = forces[2]
wy_tip = forces[3]

ax.plot([wx_tip[0], w_tip[0]],[wx_tip[1], w_tip[1]], "r--", lw=1.2)
ax.plot([wy_tip[0], w_tip[0]],[wy_tip[1], w_tip[1]], "r--", lw=1.2)

# Draw coordinate axes
x_offset, y_offset = 0.5,0.2
ax.annotate("", xy=(0.5*x_hat[0]+x_offset, 0.5*x_hat[1]+y_offset), xytext=(x_offset, y_offset),
            arrowprops=dict(arrowstyle="->", lw=1.5,alpha=0.5))
ax.text(0.6*x_hat[0]+0.4, 0.6*x_hat[1]+y_offset/2, r"$x$", fontsize=11,alpha=0.5)

ax.annotate("", xy=(0.5*y_hat[0]+x_offset, 0.5*y_hat[1]+y_offset), xytext=(x_offset, y_offset),
            arrowprops=dict(arrowstyle="->", lw=1.5,alpha=0.5))
ax.text(0.6*y_hat[0]+x_offset, 0.6*y_hat[1]+y_offset/2, r"$y$", fontsize=11,alpha=0.5)
ax.plot(x_offset,0.2,'.',color='gray',ms=10)

# Angle markers
draw_angle_arc(ax, w_dir, wy_dir, radius=0.4, label_r=0.55,label=r"$25\degree$")
draw_angle_arc(ax, wx_dir, w_dir, radius=0.25, label_r=0.35,label=r"$65\degree$")

# -----------------------------
# Proper scaling
# -----------------------------
L = max(np.linalg.norm(F) for F in forces)
ax.set_xlim(-1.4*L, 1.4*L)
ax.set_ylim(-1.4*L, 1.4*L)

plt.tight_layout()
plt.show()
`````

### Tension

A **tension** is a force along the length of a medium, such as a rope or cable.  The word "tension" comes from a Latin word meaning "to stretch".  Any flexible connecter (e.g., string, rope, wire, or cable) can exert a pull parallel to its length.  Consider the phrase: "You can't push a rope."  Instead tension force pulls outward along the two ends of a rope.

Consider a person holding a $5.0$-kg mass on a rope (see Fig. {numref}`{number}<tension-mass-fig>`).  If the mass is stationary, then its accleration is zero and the net force is zero.  The only external forces acting on the mass are its weight $w$ and the tension $T$ supplied by the rope.  Thus,

$$ \vec{F}_{\rm net} = T-w = 0, $$

where the $+\hat{j}$ direction defines up.  The tension equals the weight of the supported mass,

$$ T = w = mg, $$

hanging at rest.  For a $5.0$-kg mass (neglecting the mass of the rope), we see that

$$ T = mg = \left(5.0\ {\rm kg}\right)\left(9.81\ {\rm m/s^2}\right) = 49\ {\rm N}. $$

If we cut the rope and insert a spring, the spring would extend a length correpsonding to a force of $49\ {\rm N}$, providing a direct observation and measure of the tenstion force in the rope.

**Large Tension**

If we wish to create a large tension, we need to exert a force perpendicular to a taut flexible connector (see Fig. {numref}`{number}<tension-tightrope-fig>`).  The weight of the tightrope walker acts as a force perpendicular to the rope.

Suppose that the tightrope walker is standing still, so the net force in both directions zero.  We can decompose the tension into 2 components: right $\vec{T}_R$ and left $\vec{T}_L$. 

1. Each section of rope is supporting the weight $\vec{w}$ through their $y$ components: $T_{Ly}$ and $T_{Ry}$.  
2. The $x$ component of the weight is balanced through each section of the rope, where $T_{Lx} = T_{Rx}$.

The net force is given as

\begin{align*}
F_{{\rm net},y} &= T_{Ry} + T_{Ly} - w = 0, \\
F_{{\rm net},x} &= T_{Rx} - T_{Lx} = 0,
\end{align*} 

where we treat the tightrope walker as a particle at the center of the vectors, the $+\hat{i}$ direction goes to the right of the walker, and up is in the $+\hat{j}$ direction above the walker.

We can examine the $x$-components first, where the rope bends from the horizontal by an angle $\theta$ so that

\begin{align*}
T_{Rx} &= T_R\cos{\theta}, \\
T_{Lx} &= T_L\cos{\theta}, \\
T_R\cos{\theta} &= T_L\cos{\theta}.
\end{align*}

Since $\cos{\theta} \neq 0$, this implies $T_R = T_L$.

From the example with the hanging mass (Fig. {numref}`{number}<tension-mass-fig>`), we saw that the tension acts along the rope and is the same magnitude everywhere.  If we cut the rope then the same tension $T$ exists in along both strands, not the  sum of the forces from both sides.  As a result we can treat the left and right sides of the rope as having the same tension $T$, or

$$ \boxed{T_R = T_L = T.}$$

Turning to the $y$-components, we have $T_{Ry}$ and $T_{Ly}$ both in the positive direction and $T_{Ry} = T_{Ly} = T\sin{\theta}$ using the same logic from the $x$-component analysis.  Therefore, we can write the $y$-components as

\begin{align*}
2T\sin{\theta} - w &= 0, \\
2T\sin{\theta} &= w, \\
T &= \frac{w}{2\sin{\theta}}.
\end{align*}

Since the weight $w = mg$, we have

$$ \boxed{T = \frac{mg}{2\sin{\theta}}.}$$

We can extend this expression to describe the tension $T$ when a perpendicular force ($F_{\perp}$, or `F_{\perp}`) is exerted at the middle of a flexible connector:

$$ T = \frac{F_\perp}{2\sin{\theta}}. $$

In this case, $T$ becomes large as $\theta \to 0$.  Even the relatively small wieght of any flexible connector will cause it to sag, since an infinite tension would result if it were horizontal (i.e., $\sin{\theta}=0$).

### Friction
Friction is a resistive force the opposes motion.  The normal force acts in the opposite direction as the weight to keep an object at rest on a horizontal surface.  

- If the surface is tilted, the normal force balances the component of the weight perpendicular to the surface (i.e., $N = w_y$).  
- If the object does not slide downward, the friction balances the component of the weight parallel to the inclined plane (i.e., $f = w_x$).

### Spring Force
A spring is a special medium that has the ability to restore its shape.  To restore its shape, a spring exerts a restoring force that is proportional to and in the opposite direction in which it is stretched or compressed.

The spring force is also known as [**Hooke's law**](https://en.wikipedia.org/wiki/Hooke%27s_law), which has a mathematical form

\begin{align}
\vec{F} = -k\vec{x},
\end{align}

where $\vec{x}$ measures the position relative to the rest (relaxed) position $x_o = 0$ and the constant $k$ measures the spring's stiffness (an intrinsic property).  

The line of action is parallel to the spring axis, which means we can use the scalar form:

\begin{align}
F = -k(x-x_o),
\end{align}

where $x$ is measured along the spring axis.  We can now see that the restoring force is:

- positive when the spring is compressed; $x-x_o<0$,
- negative when the spring is compressed; $x-x_o>0$.

### Real Forces and Inertial Frames

Some forces are real, whereas others are not.  *Real* forces have some physical origin, such as gravity, tension, or friction.  Fictious forces appear only becase the observer is using an accelerating or rotating frame of reference.

For example, consider a satellite moving due north above Earth's Northern Hemisphere.  To an observer on Earth, the satellite appears to be pushed westward.  However, there is not physical force acting on the satellite in that direction.  Instead Earth is rotating eastward beneath it.  The apparent westward force exists only because Earth's frame of reference is *rotating*.

Real forces always come in action-reaction pairs.  If no physical object experiences the reaction force, then the force is fictious.

## Drawing Free-body Diagrams

THe first step in describing an analyzing most phenomena in physics involves the careful drawing of a free-body diagram (FBD).  FBDs have been used throughout this chapter.  A FBD must only include the external forces acting on the body of interest.

Once we have drawn an accurate FBD, we can apply 
- Newton's 1st law if the body is in equilibrium ($F_{\rm net} = 0$), or
- Newton's 2nd law if the body is accelerating ($F_{\rm net} \neq 0$).

```{admonition}  Constructing Free-Body Diagrams

1. Draw the object under consideration as a simpler shape (e.g., point, box, etc.).  It does not have to be artistic.  We often place the object at the origin of an $xy$-coordinate system.
2. Include all forces that act on the object, representing these forces as vectors.  Do not include the net force on the object.  We never include both forces of an action-reaction pair.
3. Convert the FBD into a more detailed diagram showing the $x$- and $y$-components of a given force.  In this case place a squiggly line through the original vector to show that it is no longer in play (i.e., replaced by components).
4. If there are more than a single object in the problem, draw a separate FBD for each object.

If there is acceleration, we do not directly include it in the FBD.  However, it may help to indicate acceleration outside the FBD.

````{exercise}
:class: etamu-exercise

**The Problem**

> Construct the free-body diagram for object A and object B in Figure {numref}`{number}<2blocks-FBD-fig>`.

<img src="https://openstax.org/apps/image-cdn/v1/f=webp/apps/archive/20251118.192121/resources/144f787db4c7c97ebb616984e0c0a811d05819b9" alt="2 blocks FBD example"  style="max-width:70%; height:auto;">

```

---

**The Model**

We consider two rigid blocks, object A and object B, in contact with each other on an inclined plane.  Object A is attached to a cord, while object B is in contact with both object A and the inclined plane.

Each object is treated as a point mass, represented by a dot.  Only forces acting *on* the isolated object are included in each free-body diagram.  Contact interactions are modeled using normal forces perpendicular to the surfaces and friction forces parallel to the surfaces.  The rope is modeled as massless and exerts a tension force along its length.

No assumptions are made here about whether the objects accelerate; the goal is only to correctly identify and label forces.

---

**The Math**

A free-body diagram is constructed by isolating a single object and drawing **all forces acting on that object**, with vectors originating at the objects center.

For **object A**, the forces are:

* Its weight $\vec{w}_A$, acting vertically downward
* The tension $\vec{T}$ from the cord, acting along the cord
* The normal force $\vec{N}_{BA}$ exerted by object B on A
* The friction force $\vec{f}_{BA}$ exerted by object B on A

For **object B**, the forces are:

* Its weight $\vec{w}_B$, acting vertically downward
* The normal force $\vec{N}_B$ exerted by the incline on B
* The friction force $\vec{f}_B$ exerted by the incline on B
* The normal force $\vec{N}_{AB}$ exerted by object A on B
* The friction force $\vec{f}_{AB}$ exerted by object A on B

Each normal force acts perpendicular to its contact surface, and each friction force acts parallel to the contact surface, opposing the tendency of relative motion.

---

**The Conclusion**

Constructing free-body diagrams requires careful attention to **which object is isolated**.  Although objects A and B exert forces on each other, those forces appear in *different* diagrams depending on the system chosen.

Drawing the diagrams separately clarifies which forces are internal to a combined system and which are external. This step is essential before applying Newtons laws to solve for motion.

---

**The Verification**

Below, we construct representative free-body diagrams for object A and object B using the same vector-drawing conventions used throughout this chapter.

`````python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Incline geometry
# -----------------------------
theta = np.deg2rad(30.0)

# Unit vectors parallel and perpendicular to incline
x_hat = np.array([np.cos(theta), np.sin(theta)])
y_hat = np.array([-np.sin(theta), np.cos(theta)])

# -----------------------------
# draw_fbd helper
# -----------------------------
def draw_fbd(ax, forces, labels, note, colors=None,offsets=None,label_offsets=None):
    ax.plot(0, 0, "b.", markersize=15)
    if colors is None:
        colors = ["black"] * len(forces)
    if offsets is None:
        offsets = [np.array([0.0, 0.0])] * len(forces)
    if label_offsets is None:
        label_offsets = [(6, 6)] * len(forces)  # default like your current
    # --- draw arrows + labels (labels use offset points so they don't rescale axes) ---
    for F, lab, col, off, laboff in zip(forces, labels, colors, offsets,label_offsets):
        Fx, Fy = F
        ox, oy = off
        ax.annotate("", xy=(Fx + ox, Fy + oy), xytext=(ox, oy),arrowprops=dict(arrowstyle="->",
                    lw=2, color=col, shrinkA=0, shrinkB=0))
        dx_pts, dy_pts = laboff
        ax.annotate(lab, xy=(Fx + ox, Fy + oy), xytext=(dx_pts, dy_pts), textcoords="offset points",
            color=col, fontsize=14, ha="center", va="bottom")
    # --- scaling: set limits from max vector length ---
    mags = [np.hypot(F[0], F[1]) for F in forces]
    R = 1.25 * max(mags + [1e-6])  # padding; avoid zero
    ax.set_xlim(-R, R)
    ax.set_ylim(-R, R)
    ax.set_aspect("equal", adjustable="box")
    ax.set_xticks([])
    ax.set_yticks([])
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.text(0.03, 0.95, note, transform=ax.transAxes, fontsize=12, va="top")
    ax.plot([-1*x_hat[0], 0], [-x_hat[1], 0],linestyle="--",color="gray",lw=1.2)
    ax.text(-0.9*x_hat[0], -1.1*x_hat[1], r"$-x$", color="gray", fontsize=11)
# -----------------------------
# Object A forces
# -----------------------------
wA = np.array([0.0, -1.0])
T  = 0.8 * x_hat
N_BA = 0.7 * y_hat
f_BA = -0.4 * x_hat
forces_A = [wA, T, N_BA, f_BA]
labels_A = [r"$\vec{w}_A$", r"$\vec{T}$", r"$\vec{N}_{BA}$", r"$\vec{f}_{BA}$"]
# Label offsets for A (tweak as you like)
label_offsets_A = [
    (0, -20),   # w_A a bit below its tip
    (0, 6),    # T slightly to the right
    (-4, 4),   # N_BA slightly left/up
    (-6, 6),  # f_BA down/left so it doesn't sit on dashed -x
]
# -----------------------------
# Object B forces
# -----------------------------
wB = np.array([0.0, -1.2])
N_B = 1.3 * y_hat
f_B = 0.5 * x_hat
N_AB = -N_BA
f_AB = -f_BA
forces_B = [wB, N_B, f_B, N_AB, f_AB]
labels_B = [r"$\vec{w}_B$",r"$\vec{N}_B$",r"$\vec{f}_B$",r"$\vec{N}_{AB}$",r"$\vec{f}_{AB}$"]

#Use offsets when two vectors point in the same direction
d = 0.12  # tweak 0.080.18 depending on your plotting scale
offsets = [np.array([0.0, 0.0]),  # wB (no offset)
    np.array([0.0, 0.0]),         # NB (no offset)
    -d/2 * y_hat,                 # fB shifted slightly along -y 
    np.array([0.0, 0.0]),         # NAB (no offset)
    +d/2 * y_hat,                 # fAB shifted slightly along +y 
]
# Label offsets for B
# Key request: make f_B label lower, and stop overlaps.
label_offsets_B = [
    (0, -20),    # w_B below
    (0, 4),    # N_B up/left a bit
    (0, -28),    # f_B clearly LOWER 
    (16, -2),    # N_AB a bit right so it doesn't sit on the arrow
    (8, 8),    # f_AB up/right so it separates from f_B label
]

# -----------------------------
# Plot
# -----------------------------
fig, axes = plt.subplots(1, 2, figsize=(9, 4), dpi=120)

draw_fbd(axes[0], forces_A, labels_A, "Object A",label_offsets=label_offsets_A)
draw_fbd(axes[1], forces_B, labels_B, "Object B",offsets=offsets,label_offsets=label_offsets_B)

plt.tight_layout()
plt.show()
`````

\vec{A}_{12} &= -\vec{A}_{21},

**The Conclusion**

The reaction-force pair at the interface is:

* On block $m_2$: $\vec{A}_{12}$ pointing to the right.
* On block $m_1$: $\vec{A}_{21}$ pointing to the left.

These two forces have the same magnitude and opposite directions, and they act on different blocks.

### **Example Problem**: Block on the Table
````{exercise}
:class: etamu-exercise

**The Problem**

> A block rests on the table, as shown in Fig. {numref}`{number}<2blocks-contact-FBD-fig>`. A light rope is attached to it and runs over a pulley. The other end of the rope is attached to a second block. The two blocks are said to be coupled. Block $m_2$ exerts a force due to its weight, which causes the system (two blocks and a string) to accelerate.
>
> We assume that the string has no mass so that we do not have to consider it as a separate object. Draw a free-body diagram for each block.

```{figure-md} 2blocks-pulley-FBD-fig
<img src="https://openstax.org/apps/image-cdn/v1/f=webp/apps/archive/20251118.192121/resources/6074c38d1454c55251b327838e603249a92fd375" alt="2 blocks pulley FBD example"  style="max-width:70%; height:auto;">

Image Credit: Openstax
```

---

**The Model**

We model the blocks as particles and the rope as a taut, massless, inextensible connector. The pulley is frictionless so the rope slides without dissipative losses. The table is taken to be frictionless (no friction force on $m_1$ is included in the free-body diagram). The rope tension has a single magnitude $T$ everywhere along the rope. The blocks share a common acceleration magnitude $a$ because the rope remains taut, with $m_1$ accelerating horizontally and $m_2$ accelerating vertically.

---

**The Math**

Apply Newtons second law separately to each block, using $+x$ to the right for $m_1$ and $+y$ upward for $m_2$.

For block $m_1$ (horizontal motion only),

\begin{align*}
\sum F_x &= m_1 a \\
T &= m_1 a .
\end{align*}

For block $m_1$ in the vertical direction (no vertical acceleration),

\begin{align*}
\sum F_y &= 0 \\
N - m_1 g &= 0 \\
N &= m_1 g .
\end{align*}

For block $m_2$ (vertical motion),

\begin{align*}
\sum F_y &= m_2 a_y \\
T - m_2 g &= m_2(-a) \\
m_2 g - T &= m_2 a .
\end{align*}

Solve the two accelerationtension equations for $a$ and $T$:

\begin{align*}
T &= m_1 a \\
m_2 g - T &= m_2 a
\end{align*}

so

\begin{align*}
m_2 g - m_1 a &= m_2 a \\
a &= \frac{m_2 g}{m_1 + m_2}, \\
T &= m_1\left(\frac{m_2 g}{m_1 + m_2}\right)\\
&= \frac{m_1 m_2 }{m_1 + m_2} g.
\end{align*}

---

**The Conclusion**

The free-body diagram for $m_1$ includes exactly three forces: the normal force $\vec{N}$ upward, the weight $\vec{w}_1 = m_1 g$ downward, and the rope tension $\vec{T}$ to the right.

The free-body diagram for $m_2$ includes exactly two forces: the rope tension $\vec{T}$ upward and the weight $\vec{w}_2 = m_2 g$ downward.

Because the rope is massless and taut over a frictionless pulley, the tension magnitude is the same on both blocks and the acceleration magnitudes are equal:

\begin{align*}
a &= \frac{m_2 g}{m_1+m_2},\\ \\
T &= \frac{m_1 m_2 g}{m_1+m_2}.
\end{align*}

---

**The Verification**

We verify the symbolic results by choosing sample values for $m_1$ and $m_2$, computing $a$ and $T$ numerically, and drawing the two free-body diagrams using those force magnitudes.

`````python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Helper: draw_fbd 
# -----------------------------
def draw_fbd(ax, forces, labels, note, colors=None, offsets=None):
    ax.plot(0, 0, "b.", markersize=15)
    if colors is None:
        colors = ["black"] * len(forces)
    if offsets is None:
        offsets = [np.array([0.0, 0.0])] * len(forces)
    for F, lab, col, off in zip(forces, labels, colors, offsets):
        Fx, Fy = F
        ox, oy = off
        ax.annotate("", xy=(Fx, Fy), xytext=(0, 0),
            arrowprops=dict(arrowstyle="->", lw=2, color=col, shrinkA=0, shrinkB=0))
        # Label a little past the tip, in display coordinates
        ax.annotate(lab, xy=(Fx + ox, Fy + oy), xytext=(8, 6),
            textcoords="offset points", color=col, fontsize=14,ha="center", va="center")
    mags = [np.hypot(F[0] + off[0], F[1] + off[1]) for F, off in zip(forces, offsets)]
    R = 1.35 * max(mags + [1.0])
    ax.set_xlim(-R, R)
    ax.set_ylim(-R, R)
    ax.set_aspect("equal", adjustable="box")
    ax.set_xticks([])
    ax.set_yticks([])
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.text(0.03, 0.95, note, transform=ax.transAxes, fontsize=12, va="top")
# -----------------------------
# Parameters (sample check)
# -----------------------------
m1 = 3.0   # kg
m2 = 2.0   # kg
g  = 9.81  # m/s^2

a = (m2 * g) / (m1 + m2)
T = m1 * a

print(f"m1 = {m1:.3g} kg")
print(f"m2 = {m2:.3g} kg")
print(f"a  = {a:.3g} m/s^2")
print(f"T  = {T:.3g} N")

# -----------------------------
# Forces for each block (scaled to real magnitudes)
# Use plot-units = N, then autoscale by draw_fbd.
# -----------------------------
# Block m1 on table
w1 = np.array([0.0, -m1*g])
N  = np.array([0.0,  m1*g])
T1 = np.array([T, 0.0])
forces_m1 = [N, w1, T1]
labels_m1 = [r"$\vec{N}$", r"$\vec{w}_1$", r"$\vec{T}$"]
colors_m1 = ["k", "k", "k"]
offsets_m1 = [[0,0],[1,0],[0,0]]
# Block m2 hanging
w2 = np.array([0.0, -m2*g])
T2 = np.array([0.0,  T])
forces_m2 = [T2, w2]
labels_m2 = [r"$\vec{T}$", r"$\vec{w}_2$"]
colors_m2 = ["k", "k"]
offsets_m2 = [[0,0],[1,0],[0,0]]

# -----------------------------
# Plot
# -----------------------------
fig, axes = plt.subplots(1, 2, figsize=(7, 3.8), dpi=120)

draw_fbd(axes[0], forces_m1, labels_m1, r"Block $m_1$", colors=colors_m1,offsets=offsets_m1)
draw_fbd(axes[1], forces_m2, labels_m2, r"Block $m_2$", colors=colors_m2,offsets=offsets_m2)

plt.tight_layout()
plt.show()
`````
